```python

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
%matplotlib inline

df = pd.read_csv("epa-sea-level.csv")
a = df["Year"]
b = df["CSIRO Adjusted Sea Level"]
plt.scatter(a,b)

slope, intercept, r, p, se = linregress(a, b)
extended_year = range(1880,2051)
y_lin_reg1 = (slope * extended_year) + intercept
plt.plot(extended_year, y_lin_reg1, c= "r")

# Create second line of best fit
xx = df[ df['Year'] >= 2000 ]['Year']
yy = df[ df['Year'] >= 2000 ]['CSIRO Adjusted Sea Level']
slope2, intercept2, r, p, se = linregress(xx, yy)
other_year = range(2000, 2051)
y_lin_reg2 = (slope2 * other_year) + intercept2
plt.plot(other_year, y_lin_reg2, c = "g")

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")

# Save plot and return data for testing (DO NOT MODIFY)
plt.savefig('sea_level_plot.png')
```
